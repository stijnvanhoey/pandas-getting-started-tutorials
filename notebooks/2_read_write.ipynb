### Read and write data from other file formats

> I want to know what other read and write methods are available in Pandas

> My colleague requested the titanic data as an spreadsheet

```python
titanic.to_excel('titanic.xlsx', sheet_name='passengers', index=False)
```

```python
titanic = pd.read_excel('titanic.xlsx', 'passengers', 
                        index_col=None, na_values=['NA'])
```
